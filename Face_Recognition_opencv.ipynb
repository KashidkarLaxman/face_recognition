# Facial Recognition  system using opencv using live trough webcam

```python
#import required modules
import cv2
import os
import numpy as np
```

```python
#detecting face from image
def face_detection(test_img):
    gray_img=cv2.cvtColor(test_img,cv2.COLOR_BGR2GRAY)
    face_haarcascade=cv2.CascadeClassifier('F:\MachineLearning/haarcascade_frontline.xml')
    faces=face_haarcascade.detectMultiScale(gray_img,scaleFactor=1.32,minNeighbors=5)
    return faces,gray_img
```

#capture images from video using opencv 

```python
# Create a VideoCapture object and read from input file
# If the input is the camera, pass 0 instead of the video file name ,we are using camera as input so pass 0
cap = cv2.VideoCapture(0)
# Check if camera opened successfully
if (cap.isOpened()== False): 
  print("Error opening video stream or file")
# Read until video is completed
i=0

while(cap.isOpened()):
  # Capture frame-by-frame
    ret, test_img = cap.read()
    face_detected,gray_img=face_detection(test_img)
    paths ='F:/MachineLearning/pictures/0/train/' + 'laxman' + str(i) + '.jpg'
    print(paths)
    cv2.imwrite(paths,test_img)
    i+=1
    
    #draw rectangle for face    
    for (x,y,w,h) in face_detected :
        cv2.rectangle(test_img,(x,y),(x+w , y+h), (255,0,0),thickness=5)
        resized_images=cv2.resize(test_img,(1000,700))
        cv2.imshow("face detcted",resized_images)
        cv2.waitKey(10)
        
        
    # Press Q on keyboard to  exitq
    if cv2.waitKey(10) == ord('q'):
        break
cap.release()

# Closes all the frames
cv2.destroyAllWindows()
```

```python
#generate training data from images and its label data
def generate_training_data():
    faces=[]
    id=[]
    for files,subfolder,folder in os.walk('F:\MachineLearning\pictures'):
        for filenames in folder:
           
            ids=os.path.basename(files)
            image_path=os.path.join(files,filenames)
            print(image_path)
            test_img=cv2.imread(image_path)
            if test_img is None:
                print("images not loaded properly")
            faces_rect,gray_img=face_detection(test_img)
            if len(faces_rect) != 1:
                continue
            (x,y,w,h)=faces_rect[0]
            roi_gray=gray_img[y:y+w,x:x+h]
            faces.append(roi_gray)
            id.append(int(ids))
        
    return faces,id
```

```python
#train classifier
def train_classifier(faces,id):
    face_recognizer=cv2.face.LBPHFaceRecognizer_create()
    face_recognizer.train(faces,np.array(id))
    return face_recognizer
#draw a rectangle on face 
def Draw_rect(test_img,face):
    (x,y,w,h)=face
    cv2.rectangle(test_img,(x,y),(x+w , y+h), (255,0,0),thickness=5)
#put name of recognized person
def put_text(test_img,text,x,y):
    cv2.putText(test_img,text,(x,y),cv2.FONT_HERSHEY_DUPLEX,5,(255,0,0),6)
```

```python
#call function for genrating training images 
faces,faceid=generate_training_data()

print(faces[0])
#Train classifer 
face_recognizer=train_classifier(faces,faceid)
#save classifier
face_recognizer.save('face_recognizer.yml')
```

```python
# Create a VideoCapture object and read from input file
# If the input is the camera, pass 0 instead of the video file name ,we are using camera as input so pass 0
names={0:'laxman','1':'sourabh',2:'basu',3:'asmita'}
cap = cv2.VideoCapture(0)
# Check if camera opened successfully
if (cap.isOpened()== False): 
  print("Error opening video stream or file")
# Read until video is completed
i=0
while(cap.isOpened()):
  # Capture frame-by-frame
    ret, test_img = cap.read()
    face_detected,gray_img=face_detection(test_img)#pass image to the face detection which will return faces detected and converted gray images
    
    for (x,y,w,h) in face_detected :
        cv2.rectangle(test_img,(x,y),(x+w , y+h), (255,0,0),thickness=5)
        resized_images=cv2.resize(test_img,(1000,700))
        cv2.imshow("face detcted",resized_images)
        cv2.waitKey(10)
    for face in face_detected:
        (x,y,w,h)=face
        roi_gray=gray_img[y:y+w,x:x+h]
        label,confidence=face_recognizer.predict(roi_gray)
        print('confidence:{},label:{}'.format(confidence,label))
        Draw_rect(test_img,face)
        if confidence <= 80:
            predicted_name=names[label]
            put_text(test_img,predicted_name,x,y)
        resized_images=cv2.resize(test_img,(1000,700))
        cv2.imshow("face detcted",resized_images)
    if cv2.waitKey(10) == ord('q'):
        break
  
cap.release()

# Closes all the frames
cv2.destroyAllWindows()
```
